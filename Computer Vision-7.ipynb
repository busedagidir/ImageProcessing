```python
# OVERLAY SMALL IMAGE ON TOP OF A LARGER IMAGE (NO BLENDING)
```

```python
img1 = cv2.imread('C:\\Users\\bused\\Desktop\\buse\\python_notlar\\computer_vision_udemy\\Computer-Vision-with-Python\\DATA\\dog_backpack.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('C:\\Users\\bused\\Desktop\\buse\\python_notlar\\computer_vision_udemy\\Computer-Vision-with-Python\\DATA\\watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2,(600,600))
```

```python
plt.imshow(img2)
```

```python
plt.imshow(img1)
```

```python
large_img = img1
small_img = img2
```

```python
x_offset, y_offset = 0,0
```

```python
x_end, y_end = x_offset + small_img.shape[1], y_offset + small_img.shape[0]
```

```python
large_img[y_offset:y_end, x_offset:x_end] = small_img 
```

```python
plt.imshow(large_img)
```
