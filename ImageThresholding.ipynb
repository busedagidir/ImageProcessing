```python
img = cv2.imread('/crossword.jpg', 0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
def show_img(img):
  fig = plt.figure(figsize=(15,15))
  ax = fig.add_subplot(111)
  ax.imshow(img,cmap='gray')
```

```python
show_img(img)
```

```python
ret, thresh2 = cv2.threshold(img,180,255,cv2.THRESH_BINARY)
show_img(thresh2)
```

```python
thresh2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11,8)
show_img(thresh2)
```

```python
blended = cv2.addWeighted(src1=img, alpha=0.6, src2=thresh2, beta=0.4, gamma=0)
show_img(blended)
```
